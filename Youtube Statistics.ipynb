```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
pd.set_option("display.float_format","{:,.3f}".format) # data type formatting to receive the result infloat with 3 decimal values
```

```python
category_subscribers_mean = df.groupby("category")["subscribers"].mean() # we found mean for the category of subscribers using groupby
category_subscribers_mean
```

```python
category_subscribers_mean.sort_values(ascending = False, inplace=True) # sort the order using sort and setting ascending= False for descending order and saving it in csv with the help of inplace=True
```

```python
category_subscribers_mean.head(10).plot(kind = "bar") # used bar chart
plt.title("Mean Subscribers By Category")
plt.xlabel("Category")
plt.ylabel("Mean Subscribers")
plt.xticks(rotation=90)
plt.show()
```

```python
channel_type_counts = df["channel_type"].value_counts() # used count aggregate function
channel_type_counts
```

```python
channel_type_counts.head(10).plot(kind = "pie", autopct= "%1.1f%%") # used pie chart
plt.title("Channel Type Distribution")
plt.show()
```

```python
plt.scatter(df["subscribers"].head(10),df["video views"].head(10)) # used scatter plot
plt.title("Scatter Plot Of Subscribers Vs Video Views")
plt.xlabel("Subscribers")
plt.ylabel("Videos View")
plt.show()
```

```python
country_subscribers_total = df.groupby("Country")["subscribers"].sum().sort_values(ascending = False) # used sum aggregate function
country_subscribers_total
```

```python
country_subscribers_total.head(10).plot(kind = "bar")
plt.title("Total Subscribers By Country")
plt.xlabel("Country")
plt.ylabel("Total Subscribers")
plt.xticks(rotation=90)
plt.show()
                                        
                                        
```

```python
df.groupby("created_year")["subscribers"].mean().sort_values(ascending = True) # used mean function
```

```python
df.groupby("channel_type")["video_views_rank"].median().sort_values(ascending =False) # median function
```

```python
df.groupby("Country")["subscribers_for_last_30_days"].max().sort_values(ascending = False) # used max aggregate function
```

```python
df.groupby("created_month")["subscribers"].sum() # applied sum aggregate function
```

```python
df.groupby("category")["video views"].std() # standard deviation function
```

```python
df.groupby("Country")["Unemployment rate"].min().sort_values(ascending = True) # used min aggregate function
```

```python
df.groupby("channel_type")["uploads"].sum() 
```

```python
df.groupby(["Latitude" , "Longitude"]).size() # For each (Latitude, Longitude) pair, it counts how many rows belong to that pair.
```

```python
sns.boxplot(x="channel_type" , y="highest_monthly_earnings", data=df,  hue="channel_type", palette="Set2",dodge=False) # hue adds color differentiation by 'channel_type'
    # palette="Set2",  Specifies the color palette
   # dodge=False   Ensures all boxplots for the same category are stacked together
plt.title("Distribution of Highest Monthly Earning by Channel Type")
plt.xticks(rotation = 90)
plt.legend([], [], frameon=False) # Hiding the legend (can be customized or removed entirely if unnecessary)
plt.show() # Its a boxplot
```

```python
df.groupby(["category" , "Country"])["subscribers"].sum().unstack().plot(kind = "bar", stacked = True) # unstack make a column format
plt.title("Subscribers Breakdown By Category and Country")
plt.xlabel("Category")
plt.ylabel("Total Subscribers")
plt.legend(title = "Country" , bbox_to_anchor = (1.05, 1), loc = "upper left")
plt.xticks(rotation = 90)
plt.show() # stacked bar chart
```

```python
plt.scatter(df["Gross tertiary education enrollment (%)"],df["subscribers"], alpha= 0.5) # uses as a parameter to show the transparency level of visual
plt.title("Scatter Plot Of Subscribers Vs Gross Tertiary Education Enrollment")
plt.xlabel("Gross Tertiary Education Enrollment")
plt.ylabel("Subscribers")
plt.show()
```

```python
df.groupby("channel_type")["Unemployment rate"].mean().sort_values(ascending = False).plot(kind = "bar")
plt.title("Average Unemployment Rate By Channel Type")
plt.xlabel("Channel Type")
plt.ylabel("Average Unemployment Rate")
plt.xticks(rotation=90)
plt.show()
```

```python
df.groupby(["created_year","channel_type"])["subscribers"].mean().unstack().tail().plot(kind = "bar" , width = 1) # multicoulmn
plt.title("Average Subscribers By Created Year And Channel Type")
plt.xlabel("Created Year")
plt.ylabel("Average_Subscribers")
plt.xticks(rotation = 45)
plt.legend(title = "Channel Type",bbox_to_anchor = (1.05,1), loc="upper left")
plt.show()
```

```python
df.groupby(["created_year" , "created_month"])["subscribers"].sum().unstack().plot(kind = "area", stacked = True)
plt.title("Subscribers Growth Over Time")
plt.xlabel("Year")
plt.ylabel("Total Subscribers")
plt.legend(title = "Month", bbox_to_anchor = (1.05,1), loc = "upper left")
plt.xticks(rotation = 45)
plt.show()
```

```python
df.groupby(["created_year" , "Country"])["highest_yearly_earnings"].mean().unstack().head().plot(kind = "line" ,marker="o")
plt.title("Trend Of Highest Yearly Earnings by Created Year and Country")
plt.xlabel("Created Year")
plt.ylabel("Average Highest Yearly Earnings")
plt.legend(title = "Country", bbox_to_anchor =(1.05,1) , loc = "upper left")
plt.xticks(rotation= 45)
plt.show()
```

```python
category_highest_monthly_earnings_sum = df.groupby("category")["highest_monthly_earnings"].sum().sort_values(ascending= False )
category_highest_monthly_earnings_sum
```

```python
category_highest_monthly_earnings_sum.head().plot(kind = "pie", autopct= "%1.1f%%")
plt.title("Percentage Of Highest Monthly Earnings By Category")
plt.ylabel("")
plt.show()
```

```python
channel_created_subscribers_mean = df.groupby(["channel_type","created_year"])["subscribers_for_last_30_days"].mean().unstack()
```

```python
df.groupby("created_month")["subscribers_for_last_30_days"].mean().plot(kind="line",marker="o")
plt.title("Average Subscribers for Last 30 Days by Month")
plt.xlabel("Month")
plt.ylabel("Average Subscribers for Last 30 Days")
plt.xticks(rotation = 45)
plt.show()
```
